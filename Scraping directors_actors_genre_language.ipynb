```python
import pandas as pd
import numpy as np
import imdb
```

```python
df = pd.read_csv("Movie_data_id.csv")
```

```python
df.head()
```

```python
imdb_list = list(df['IMDB_ID'])
```

```python
imdb_movie = imdb.IMDb()
directors_list = []
actors_list = []
genres_list = []
languages_list = []

for movie_id in imdb_list:
    
    if movie_id == 'No ID found':
        
        directors_list.append('NAN')
        actors_list.append('NAN')
        genres_list.append('NAN')
        languages_list.append('NAN')
        
    else:
        
        movie_obj = imdb_movie.get_movie(movie_id)

        try:
            directors = []
            for director in movie_obj['directors']:
                directors.append(str(director))
            directors = ", ".join(directors)
            
        except:
            directors = 'NAN'

        try:
            actors = []
            for actor in movie_obj['cast']:
                actors.append(str(actor))
            actors = ", ".join(actors)
            
        except:
            actors = 'NAN'
            
        try:
            genres = []
            for genre in movie_obj['genres']:
                genres.append(str(genre))
            genres = ", ".join(genres)
                
        except:
            genres = 'NAN'
            
        try:
            languages = []
            for language in movie_obj['languages']:
                languages.append(str(language))
            languages = ", ".join(languages)
                
        except:
            languages = 'NAN'
        

        directors_list.append(directors)
        actors_list.append(actors)
        genres_list.append(genres)
        languages_list.append(languages)
```

```python
df['Directors'] = directors_list
df['Actors'] = actors_list
df['Genres'] = genres_list
df['Languages'] = languages_list
df.to_csv('Dataset.csv',encoding='utf-8-sig',index=False)
```
